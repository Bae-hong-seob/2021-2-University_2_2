```python
'''
가설 : 주택의 가격과 출산율이 음의 상관관계가 있다 

개요
- 우리나라의 합계출산율은 2018년 출산율이 0.98명을 기록하면서
  가임기 여성이 평균적으로 출산할 것으로 기대되는 자녀의 수가
  우리나라 합계출산율 집계 이후 처음으로 1명에 미치지 않게 되었다.
  
- 또한 출산을 함으로써 얻게 되는 효용이 출산으로 발생한 수 있는 비용보다 
  충분히 크지 않다면 자녀 출산을 기피할 가능성이 높고 그 중 주택구입 비용은 
  한 가정이 지출하게 되는 가장 큰 목돈에 해당하기 때문에 더욱 출산율이 낮아지고 있다고 생각

- 마지막으로 이러한 주택의 가격 변화가 사람들의 거주공간 이동에 영향을 미칠것이라고 생각
  
사욜할 데이터
- 주택매매가격지수
- 주택전세가격지수
- 아파트 실거래가격지수
- 서울시 연월별 출산율
- 지역별 인구 이동건수
'''
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
hs = pd.read_csv("주택매매가격지수(KB).csv")
hl = pd.read_csv("주택전세가격지수(KB).csv")
ap = pd.read_csv("아파트 실거래가격지수.csv")
```

```python
hs.shape
```

```python
type(hs)
type(hl)
```

```python
hs.head()
```

```python
hs.plot(x='Unnamed: 0', figsize=(16, 5), grid=True)
plt.title("주택매매가격지수")
plt.rc('font', family='Malgun Gothic')
plt.show()
#모든 유형의 주택들의 매매가격지수가 과거부터 조금씩 증가하는것을 확인
```

```python
hl.plot(x='Unnamed: 0', figsize=(16, 5), grid=True)
plt.title("주택전세가격지수")
plt.rc('font', family='Malgun Gothic')
plt.show()
#주택매매가격지수와 마찬가지로 전세가격지수도 거의 유사한 그래프를 가진다.
```

```python
bir = pd.read_csv("시군구_성_월별_출생.csv",encoding='cp949') #kosis에서 가져온 전국 지역별 출산정보 데이터
bir.head()
```

```python
bir.set_index('Unnamed: 0', inplace = True) #인덱스를 지역명(Unnamed: 0)으로 바꿈
bir.head()
```

```python
bir.info()
```

```python
bir_dt = bir.loc[:, "2010. 01":"2020. 12"] #지역별 출산율 데이터 중 2010-2020년 데이터만 추출
bir_dt = bir_dt.transpose() #앞서 살펴본 주택 매매,전세 가격지수와 비교하기 위해 출산율 데이터의 행열 변환
bir_dt.head()
```

```python
#전국 지역 중 서울을 대상으로 분석을 실행하기 위해 열 이름 '서울특별시' 따로 추출
bir_dt['서울특별시'].unique() 
```

```python
bir_dt['서울특별시'] = bir_dt['서울특별시'].astype('int32') 
#서울특별시의 데이터 타입 확인 결과 'object'인것을 확인하여 정수형으로 변환
```

```python
bir_dt.plot(y=['서울특별시'], figsize=(25, 15), grid=True)
plt.title("서울특별시 출산율 그래프")
plt.rc('font', family='Malgun Gothic')
plt.show()
# 서울특별시의 출산율이 과거부터 시간의 흐름에 따라 감소하는것을 확인
```

```python
hs_year = hs.loc[288:419] #주택매매가격지수(2010~2020)
hs_year.head()
```

```python
meme = hs_year.plot(x='Unnamed: 0', y='총지수(서울)[2019.01=100]', figsize=(16, 6))
plt.title("서울특별시의 주택매매가격지수와 출산율")
plt.xlabel("연도")
plt.ylabel("주택매매가격")
bir_dt.plot(y='서울특별시', secondary_y=True, ax=meme)
plt.rc('font', family='Malgun Gothic')
plt.show()
# 서울의 출산율과 주택매매가격이 서로 반비례(음의 상관관계)를 확인할 수 있었음
```

```python
hl_year = hl.loc[288:419] #주택전세가격지수(2010~2020) 추출
hl_year.head()
```

```python
junsae = hl_year.plot(x='Unnamed: 0', y='총지수(서울)[2019.01=100]', figsize=(16, 6))
plt.title("서울특별시의 주택전세가격지수와 출산율")
plt.xlabel("연도")
plt.ylabel("주택전세가격")
bir_dt.plot(y='서울특별시', secondary_y=True, ax=junsae)
plt.rc('font', family='Malgun Gothic')
plt.show()
# 서울의 출산율과 주택전세가격 또한 앞서 살펴본 주택매매가격지수와 마찬가지로 출산율과 음의 상관관계를 가짐을 확인할 수 있었음
```

```python
'''
주택가격과 출산율의 관계를 서울특별시에 한해서만 분석하였기 때문에 상관관계 입증이 어렵다고 생각
따라서 부산광역시의 아파트실거래가 데이터 시각화하여 출산율 데이터와 비교
'''
```

```python
bir_dt['부산광역시'] = bir_dt['부산광역시'].astype('int32') #부산광역시의 출산율 데이터 역시 자료형 변환

bir_dt.plot(y=['부산광역시'], figsize=(25, 15), grid=True)
plt.title("부산광역시 출산율 그래프")
plt.rc('font', family='Malgun Gothic')
plt.show()
```

```python
apart = ap_year.plot(x='Unnamed: 0', y='부산[2017.11=100]', figsize=(16, 6))
plt.title("부산광역시 아파트 실거래가와 출산율")
plt.xlabel("연도")
plt.ylabel("아파트 실거래가")
bir_dt.plot(y='부산광역시', secondary_y=True, ax=apart)
plt.rc('font', family='Malgun Gothic')
plt.show()
#부산광역시의 아파트 실거래가와 부산지역의 출산율을 비교하였을때, 위와 마찬가지로 반비례 관계를 확인할 수 있다.
```

```python
pm = pd.read_csv("시군구별_이동건수.csv",encoding='cp949')
pm.head()
```

```python
pm = pm.drop(["이동규모별"], axis=1) #필요없는 열 삭제(전처리)
pm.head()
```

```python
pm.info()
```

```python
pm.set_index("행정구역(시군구)별",inplace = True)
pm.astype('float')
```

```python
pm = pm.transpose()
pm.head()
```

```python
plt.boxplot(pm)
plt.title("서울 인구이동 박스플롯")
plt.xticks([1,8], ["서울","경기도"])
#전국적으로 경기도가 유일하게 전입인원이 +이다
```

```python
plt.boxplot(pm["서울특별시"],sym="b*")
plt.title("서울 인구이동 박스플롯")
plt.xticks([1], ["서울 순이동 인구"])
plt.show()
#서울의 인구 이동을 보았을때, 인구가 대부분 전출하는 것을 알 수 있다.
```

```python
#서울특별시의 인구이동에 따라서 서울주택 매매,전세가격의 차이를 알아보려 한다.
pm.plot(y=['서울특별시'], figsize=(25, 9), grid=True)
plt.title("서울특별시 이동 인구")
plt.rc('font', family='Malgun Gothic')
plt.show()
```
